```python
# Подключение необходимых библиотек
import numpy as np
import pandas as pd
import math
from datetime import datetime, timedelta, date
import matplotlib.pyplot as plt

# ОТКРЫТИЕ ВСЕХ ДАТАСЕТОВ
df_students = pd.read_csv("students.csv", parse_dates=['birthday'], dayfirst=False)
df_students.rename(columns={'id': 'student_id'}, inplace=True)
del df_students['id_']

df_courses = pd.read_csv("courses.csv", index_col=0)
df_courses.rename(columns={'id': 'course_id'}, inplace=True)

df_course_contents = pd.read_csv("course_contents.csv")

df_progresses = pd.read_csv("progresses.csv")
df_progresses.rename(columns={'id': 'progress_id'}, inplace=True)

df_progress_phases = pd.read_csv("progress_phases.csv", 
                                 parse_dates=['start_date', 'finish_date'], dayfirst=False)

# ОБЪЕДИНЕНИЕ ДАТАСЕТОВ
# Объединяем датасеты progress_phases и progresses
# df_main = df_progress_phases.merge(df_progresses, left_on='progress_id', right_on='id', how='left')
df_main = df_progress_phases.merge(df_progresses, on='progress_id', how='outer')
# Объединяем с датасетом course_contents
df_main = df_main.merge(df_course_contents, on=['course_id', 'module_number', 'lesson_number'], how='outer') #)'outer'right
# df_main = df_main.merge(df_courses, left_on='course_id', right_on='id', how='left')
df_main = df_main.merge(df_courses, on='course_id', how='left') #)'outer'
# Изменим для удобства имя столбца с названием курса
df_main.rename(columns={'title': 'course_title'}, inplace=True)
# Добавим столбец счетчик
df_main['N'] = 1
```

```python
# МИНИМАЛЬНЫЙ, МАКСИМАЛЬНЫЙ, СРЕДНИЙ, МЕДИАННЫЙ ВОЗРАСТ СТУДЕНТОВ
# Функция, возвращающая возраст, либо 0
def calculateAge(birth_Date):
    try:
        i = int((datetime.today() - datetime.strptime(birth_Date, '%Y-%m-%d')).days / 365.2425)
    except Exception as exc:
        i = 0
    return i

# Копируем датасет данных о студентах в датасет для подсчета их возрастов
df_students_age = df_students.copy()
# Определяем возраст студентов
df_students_age['age'] = df_students_age['birthday'].apply(calculateAge)
# Удаляем некорректные данные
df_students_age = df_students_age[df_students_age['age'] != 0]
# Печатаем требуемые параметры
print('Минимальный возраст студентов:  ', df_students_age['age'].min())
print('Максимальный возраст студентов: ', df_students_age['age'].max())
print('Средний возраст студентов:      ', df_students_age['age'].mean())
print('Медианный возраст студентов:    ', df_students_age['age'].median())

# Оценим количество некорректных данных
print("Количество записей с пустыми датами рождения  - ", len(df_students.index) - len(df_students_age.index))
print("Количество записей с возрастом 10 и менее лет - ", len(df_students_age[(df_students_age['age'] < 10)].index) )
print("Количество записей с возрастом более 100 лет  - ", len(df_students_age[(df_students_age['age'] > 100)].index) )
```

```python
# Скорректируем данные о возрастах, оставив в датасете только строки с возрастами от 10 до 100
df_students_age = df_students_age[(10 <= df_students_age['age']) & (df_students_age['age'] <= 100)]

# Печатаем требуемые параметры
print('Минимальный возраст студентов:  ', df_students_age['age'].min())
print('Максимальный возраст студентов: ', df_students_age['age'].max())
print('Средний возраст студентов:      ', df_students_age['age'].mean())
print('Медианный возраст студентов:    ', df_students_age['age'].median())
```

```python
# МИНИМАЛЬНЫЙ, МАКСИМАЛЬНЫЙ, СРЕДНИЙ, МЕДИАННЫЙ ВОЗРАСТ СТУДЕНТОВ НА КАЖДОМ КУРСЕ
# Объединяем датасеты students_age, progresses и courses
df_main_students = df_progresses.merge(df_students_age, on='student_id', how='inner')
df_main_students = df_main_students.merge(df_courses, on='course_id', how='left')
df_main_students.rename(columns={'title': 'course_title'}, inplace=True)

df_main_students.groupby('course_title')['age'].agg({'min', 'max', 'mean', 'median'})
```

```python
# ПУНКТ 1.3. 
# Строим горизонтальный bar-chart, отражающий количество студентов на каждом курсе.
df = pd.DataFrame(df_main.groupby('course_title')['student_id', 'field'].agg({'student_id':'nunique', 'field':'first'}))
df.reset_index(inplace=True)
df.sort_values('student_id', inplace = True)

# color_map = {'Marketing':'red', 'Design':'blue', 'Development':'green', 'Business':'magenta'}
# df['color'] = df['field'].apply(lambda x: color_map[x])

plt.figure(figsize=(10, 5))

for name, row in df.groupby(['field']):
    plt.barh(row['course_title'], row['student_id'], alpha=0.1, label=name)

median = df['student_id'].median()
plt.axvline(x=median, color='purple', alpha=0.3)

plt.title('Количество студентов на каждом курсе')
plt.legend()
plt.box(on=None)
plt.show()
```

## 4. Расчет конверсии

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).
2. **(m)** Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы
3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
# ПУНКТ 4.1
# Посчитаем конверсию перехода студентов из одного модуля в другой на каждом курсе.

# Подсчитаем по каждому курсу и модулю количество студентов, начавших модуль
# Используем данные только по домашнему заданию
df = df_main[df_main['is_homework'] == True]
# Сгруппируем данные по кусру и модулю, подсчитав количество студентов. Добавим данные о сфере курса и названию модуля
df = pd.DataFrame(df.groupby(['course_title', 'module_number'])['field', 'module_title', 'student_id'].\
                  agg({'field':'first', 'module_title':'first', 'student_id':'count'})).reset_index()
# Переименуем столбец для удобства
df.rename(columns={'student_id': 'count_of_start'}, inplace=True)

# Подсчитаем по каждому курсу и модулю количество студентов, закончивших модуль
# Используем данные только по домашнему заданию и со статусом 'done'
df1 = df_main[(df_main['is_homework'] == True) & (df_main['status'] == 'done')]
# Сгруппируем данные по кусру и модулю, подсчитав количество студентов
df1 = pd.DataFrame(df1.groupby(['course_title', 'module_number'])['student_id'].\
                  agg('count')).reset_index()
# Переименуем столбец для удобства
df1.rename(columns={'student_id': 'count_of_finished'}, inplace=True)

# Объединим оба полученных датафрейма
df = df.merge(df1, on=['course_title', 'module_number'], how='left')

# Подсчитаем конверсию
conversion = []
conv_proc = []
prev_course = df.iloc[0]['course_title']
prev_count = df.iloc[0]['count_of_start']
first_count = df.iloc[0]['count_of_start']
# Для каждой строки в подготовленной выше таблице
for i, row in df.iterrows():
    if row['course_title'] == prev_course:
        # Для строк внутри курса - сравниваем с данными первого или предыдущего модуля
        conversion.append( prev_count )
        prev_count = row['count_of_finished']
    else:
        # При переходе к новому курсу - сохраняем даные о первом модуле курса
        prev_course = row['course_title']
        prev_count = row['count_of_finished']
        first_count = row['count_of_finished']
        conversion.append( row['count_of_start'] )
    conv_proc.append( round( row['count_of_start'] / first_count * 100,2))

# Запишем значения конверсии в новые столбцы
df['conversion'] = df['count_of_start'] / conversion
df['conv_proc'] = conv_proc 

df.head()
```

## 5. Метрика успеваемости 

&nbsp;&nbsp;&nbsp;&nbsp;*(необязательное задание)*

Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

*Предлагается определить метрику оценки успеваемости студента как сумму двух отношений:*

- *отношение медианного времени, затраченного студентом на прохождение уроков, к аналогичному показателю по всем студентам курса;*
- *отношение количества незаконченных уроков к общему их количеству на курсе.*

*Очевидно, чем выше данный показатель, тем хуже успеваемость студента.*

*Кроме того, детально оценивать причины плохой успеваемости можно и отдельно по каждой из составляющих.*

```python
# Создадим копию основной базы и посчитаем по каждому студенту время, потраченное на прохождение урока 
df_rating = df_main.copy()
df_rating['term'] = (df_rating['finish_date'] - df_rating['start_date']).dt.seconds
df_rating['term'] = df_rating['term'].fillna(-1)     # Пометим пустые значения
df_rating = df_rating[df_rating['term'] != -1]       # И уберем их

# Создадим датафрейм, в котором по каждому курсу укажем количество уроков и медианное время прохождения урока
# Считаем медианное время, потраченное на прохождение уроков всеми студентами курса
df = pd.DataFrame(df_rating.groupby('course_id')['term'].agg('median'))
# Добавляем количество уроков в курсе
df = df.merge(pd.DataFrame(df_course_contents.groupby('course_id')['lesson_number'].agg('count')), on='course_id', how='left')
# Изменим для удобства имена столбцов
df.rename(columns={'term': 'course_median_term', 'lesson_number': 'course_lesson_count'}, inplace=True)

# По каждому студенту считаем медианное время, потраченное на прохождение урока и количество пройденных уроков
df_rating = pd.DataFrame(df_rating.groupby(['student_id', 'course_id'])['term', 'lesson_number'].\
                         agg({'term':'median', 'lesson_number':'count'})).reset_index()
# Изменим для удобства имена столбцов
df_rating.rename(columns={'term': 'median_term', 'lesson_number': 'lesson_count'}, inplace=True)
# Объединяем две таблицы
df_rating = df_rating.merge(df, on='course_id', how='left')

# Считаем отношение числа пройденных уроков к общему их количеству на курсе
df_rating['ratio_term'] = df_rating['median_term'] / df_rating['course_median_term']
# Считаем отношение медианного времени прохождения урока студентом к аналогичному параметру по курсу
df_rating['ratio_lesson'] = df_rating['lesson_count'] / df_rating['course_lesson_count']
# Считаем сумму отношений, записываем в датасет
df_rating['rating'] = df_rating['ratio_term'] + df_rating['ratio_lesson']

# Сортируем по убыванию рейтинга
df_rating.sort_values('rating', ascending = False, inplace = True)

# Выведем основные метрики нашего нового параметра rating
print('Минимум      : ', round(df_rating['rating'].min(), 2))
print('Максимум     : ', round(df_rating['rating'].max(), 2))
print('Среднее      : ', round(df_rating['rating'].mean(), 2))
print('Медиана      : ', round(df_rating['rating'].median(), 2))
print('99% квантиль : ', round(df_rating['rating'].quantile(0.99), 2))

# Посторим гистограмму распределения параметра rating
plt.hist(df_rating['rating'], bins=100)
plt.show()
```
